# Import Libraries and Dataset

```python
import pandas as pd
import numpy as np
movies = pd.read_csv("movies.csv")
pd.set_option('display.max_columns', None)
movies.head(100)
```

### here we have gathered the 'genres','keywords','cast','director' into a new attribute name features

```python
features = ['genres','keywords','cast','director']
```

### helper Fuction

```python
def combine_feature(row):
    return row["genres"]+" "+row["cast"]+" "+row["keywords"]+" "+row["director"]
```

```python
# we need to fill any na value in the selected attribute values with '' and than apply the helper function
for feature in features:
    movies[feature] = movies[feature].fillna('')
movies["combined_feature"] = movies.apply(combine_feature,axis = 1)
```

```python
movies.head()
```

## tokenize the combined_feature using  CountVectorizer which will return the sparse matrix

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv= CountVectorizer()
count_matrix = cv.fit_transform(movies["combined_feature"])
count_matrix
```

## The cosine_similarity is used to find out the which movies are similar to which one and a final matrix is formed


<img src = "cosine_example.png" alt = "alterante text" width="200" align = "left"/> 

```python
from sklearn.metrics.pairwise import cosine_similarity
cos_similarity = cosine_similarity(count_matrix)
print(cos_similarity)
```

```python
def get_title_from_index(index):
    return movies[movies.index == index]["title"].values[0]
def get_index_from_title(title):
    return movies[movies.title == title]["index"].values[0]
def get_releaseDate_from_index(index):
    return movies[movies.index == index]["release_date"].values[0]
```

## Using tkinter a UI is been created in which if reuired input is given than the correct movies will be represented

```python
#!/usr/bin/python
# -*- coding: utf-8 -*-

import os
import os.path
import sys
import tkinter as tk
from tkinter import filedialog
from tkinter import messagebox
from tkinter import *
from tkinter import ttk
import shutil
import threading
from pathlib import Path


class MainWindow:

#     THIS_FOLDER_G = ''
#     if getattr(sys, 'frozen', False):
#         THIS_FOLDER_G = os.path.dirname(sys.executable)
#     else:
#         THIS_FOLDER_G = os.path.dirname(os.path.realpath(__file__))

    def __init__(self, root):
        self.root = root
        self._inp = tk.StringVar()
        self._status = tk.StringVar()

        root.title('Movie Recommender')
        root.configure(bg='#eeeeee')

        self.inp_label = tk.Label(root, text='Enter the Movie Name:',
                                  bg='#eeeeee', anchor=tk.W)
        self.inp_label.grid(
            padx=50,
            pady=(8, 0),
            ipadx=2,
            ipady=1,
            row=0,
            column=2,
            columnspan=8,
            sticky=tk.W + tk.E + tk.N + tk.S,
            )

        self.inp_entry = tk.Entry(root, textvariable=self._inp,
                                  bg='#fff', width = "70",exportselection=0,
                                  relief=tk.FLAT)
        self.inp_entry.grid(
            padx=15,
            pady=6,
            ipadx=8,
            ipady=8,
            row=1,
            column=0,
            columnspan=8,
            sticky=tk.W + tk.E + tk.N + tk.S,
            )

        self.recommend_btn = tk.Button(
            root,
            text='RECOMMEND',
            command=self.recommend_callback,
            bg='#27ae60',
            fg='#ffffff',
            bd=2,
            relief=tk.FLAT,
            )
        self.recommend_btn.grid(
            padx=15,
            pady=6,
            ipadx=24,
            ipady=6,
            row=8,
            column=0,
            columnspan=8,
            sticky=tk.W + tk.E + tk.N + tk.S,
            )

        self.status_label = tk.Label(
            root,
            textvariable=self._status,
            bg='#eeeeee',
            anchor=tk.W,
            justify=tk.LEFT,
            relief=tk.FLAT,
            wraplength=350,
            )
        self.status_label.grid(
            padx=12,
            pady=(0, 12),
            ipadx=0,
            ipady=1,
            row=9,
            column=0,
            columnspan=8,
            sticky=tk.W + tk.E + tk.N + tk.S,
            )

    def recommend_callback(self):
        if len(self._inp.get()) == 0:
            messagebox.showinfo('MR', 'Please Enter a valid Input !!')
            return
        t1 = threading.Thread(target=self.recommend, daemon=True)
        t1.start()

    def recommend(self):
        try:
            # take the input of the movie name strictly their original name for example Iron Man or Spider-Man
            # from the movie title we need to find the movie index to see simmilar movies in simmilarity matrics
            # sort them in descending order
            movie_user_likes = str(self._inp.get())
            movie_index = get_index_from_title(movie_user_likes)
            similar_movies = list(enumerate(cos_similarity[movie_index]))
            sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]

            i=0
            ans = []
            for element in sorted_similar_movies:
                st = str(get_title_from_index(element[0]))+" {"+str(get_releaseDate_from_index(element[0])[0:4])+"}"
                ans.append(st)

                i=i+1
                if i>5:
                    break
            temp2 = "\n".join(ans)
            s = "Top 5 similar movies to "+movie_user_likes+" are:\n"
            self._status.set(s+temp2)
            self.status_label.update()
        except Exception as e:
            self._status.set("Enter valid movie name")
            self.status_label.update()
ROOT = tk.Tk()
ROOT.geometry("500x500")
MAIN_WINDOW = MainWindow(ROOT)
ROOT.mainloop()
```
